# Hearing the phase of a sound

We will investigate the effect of phase on the perceptual quality of a sound. The human ear is largely insensitive to phase and that's why most of the equalization in commercial-grade audio equipment takes place in the magnitude domain only.

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, ifft
from IPython.display import Audio
import IPython
from scipy.io import wavfile
```

```python
plt.rcParams["figure.figsize"] = (15,6)
```

We will be synthesizing audio clips so let's set the sampling rate for the rest of the not:

```python
Fs = 16000
TWOPI = 2 * np.pi
```

We will be synthesizing and playing audio clips so define a convenience function to "beautify" the resulting sound: avoid abrupt "clicks" when the waveform begins and ends. 

```python
def prepare(x, max_value=3):
    N = len(x)
    # fade-in and fade-out times max 0.2s
    tf = min(int(Fs*0.2), int(N*0.1))
    for n in range(int(tf)):
        s = float(n) / float(tf)
        x[n] *= s
        x[N-n-1] *= s
    x = np.concatenate((x, np.linspace(0, max_value, int(Fs/2)), np.linspace(max_value, 0, int(Fs/2))))
    return x
```

## 1) Sustained sounds

The first experiment will use sustained sounds, i.e. waveform whose global envelope does not change in time. A periodic sustained waveform is simply the sum of harmonically-related sinusoidal components, i.e. a set of sines or cosines with different amplitudes and phases at frequencies multiple of a fundamental. The fundamental frequency is also known as the <i>pitch</i> of the sound.

A class of instruments that produce good sustained sounds is the woodwinds (clarinet, saxophone, etc). For our purposes we will choose to simulate a clarinet according to the analysis described [here](http://www.phy.mtu.edu/~suits/clarinet.html).

A clarinet-like sustained sound will contain frequencies at odd multiples of the fundamental. We will just keep the fundamental and five harmonics and we be able to specify the initial phase offset for each component:

```python
def clarinet(f, phase=[]):
    # length of audio clip
    T = 3
    
    # we will keep 5 harmonics and the fundamental amplitude of components: 
    ha = [0.75, 0.5, 0.14, 0.5, 0.12, 0.17]
    
    # phase
    phase = np.concatenate((phase, np.zeros(len(ha) - len(phase))))
    
    x = np.zeros((T*Fs))
    # clarinet has only odd harmonics
    n = np.arange(len(x))
    for k, h in enumerate(ha):
        x += h * np.sin(phase[k] + TWOPI * (2*k + 1) * (float(f)/Fs) * n)
    return x
```

```python
# fundamental frequence: D4
D4 = 293.665
x = clarinet(D4)

plt.plot(x[:300]);
plt.show();

Audio(prepare(x), rate=Fs)
```

```python
xrp = clarinet(D4, [3.84, 0.90, 3.98, 4.50, 4.80, 2.96])

plt.plot(xrp[:300]);
plt.show();

# Audio sounds the same
Audio(prepare(xrp), rate=Fs)
```

So it seems that phase is not important after all. To check once again, run the following notebook cell as many times as you want and see if you can tell the difference between the original zero-phase and a random-phase sustained note (the phases will be different every time you run the cell):

```python
xrp = clarinet(D4, np.random.rand(6)*TWOPI)
plt.plot(xrp[:300]);
plt.show();
IPython.display.display(Audio(prepare(x), rate=Fs))
IPython.display.display(Audio(prepare(xrp), rate=Fs))
```

## 1) Dynamic sounds

In the second experiment we will use real-world dynamic sounds, i.e. sounds that display time-varying characteristics. Typically, a physical musical instrument will produce sounds whose envelope displays four subsequent portions:

* the **attack** time is the time taken for the sound to go from silence to max amplitude
* the **decay** time is the time taken for the sound to decrease to sustain level
* the **sustain** time is the time during the sound is kept at the same amplitude
* the **release** time is the time taken for sound to go to zero after the stimulation is stopped.

Consider for instance a piano note: the attack time is very quick (the hammer hits the string); the decay is quite rapid as the string settles into harmonic equilibrium but there is no sustain since once the hammer hits, the stimulation ends. So a piano note has a distinct volume envelope that rises very fast and then releases slowly:

```python
Fs, x = wavfile.read("piano.wav")
plt.plot(x);
plt.show();
Audio(x, rate=Fs)
```

The shape of a waveform is largely encoded in the phase. Therefore if we mess up with the phase of the piano sample above, we will get something that looks very different.

To see this, let's take the DFT of the audio data, set the phase to zero and take the IDFT:

```python
# prep work
if len(x) % 2 != 0:
    x = x[:-1]

mv = int(max(abs(x)) * 1.2)
```

```python
X = fft(x)

plt.plot(np.abs(X[0:int(len(X)/2)]));
plt.show();
```

```python
# to set phase=0 we need to take magnitude of DFT
xzp = ifft(np.abs(X))

# Theoretically, xzp should be real
# Due to numerical imprecision, we are left with imag value
print(max(np.imag(xzp)) / max(abs(xzp)))
```

```python
# imag part is negligible
xzp = np.real(xzp)

plt.plot(xzp);
plt.show();
```

By removing the phase, we have destroyed the timing information that, for instance, made the sharp attack possible (mathematically, note that by creating a zero-phase spectrum we did obtain a symmetric signal in the time domain!).

If we play the waveform, we can hear that the pitch and some of the timbral quality have been preserved (after all, the magnitude spectrum is the same), but the typical piano-like envelope has been lost.

```python
Audio(prepare(xzp, mv), rate=Fs)
```

Let's set a random phase for each DFT component. The only tricky thing here is that we need to preserve the Hermitian symmetry of the DFT in order to have a real-valued time-domain signal:

```python
# we know the signal is even-length so we need to build
#  a phase vector of the form [0 p1 p2 ... pM -pM ... -p2 -p1]
#  where M = len(x)/2
ph = np.random.rand(int(len(x)/2) ) * TWOPI * 1j
ph = np.concatenate(([0], ph, -ph[-2::-1]))

xrp = ifft(X * np.exp(ph))

print(max(np.imag(xrp)) / max(np.abs(xrp)))
```

```python
xrp = np.real(xrp)

plt.plot(xrp);
plt.show();
Audio(prepare(xrp, mv), rate=Fs)
```
