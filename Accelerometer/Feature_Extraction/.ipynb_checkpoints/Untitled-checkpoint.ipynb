```python
df.plot(y="Absolute acceleration (m/s^2)");
```

```python
fs = 1/0.010073
fhp = signal.cheby2(N=20, rs=180, Wn=3, fs=fs, btype='hp', output='sos')
```

```python
df['filtered_x'] = signal.sosfilt(fhp, df['Linear Acceleration x (m/s^2)'])
df.plot(y=["Linear Acceleration x (m/s^2)", "filtered_x"]);
```

```python
df['filtered_abs'] = signal.sosfilt(fhp, df['Absolute acceleration (m/s^2)'])
df.plot(y=["Absolute acceleration (m/s^2)", "filtered_abs"]);
```

```python
freqs, psd = signal.welch(df['filtered_abs'], fs=fs)

plt.semilogx(freqs, psd)
plt.title('PSD: power spectral density')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.tight_layout()
```
